## Use of conditions

```python
# Check which rows in the 'firstName' column are NOT equal to "A" (!=)
# This returns a Boolean Series (True = not "A", False = "A")

dataframe["firstName"] != "A"
```

```python
# Filter the DataFrame to include only rows where 'firstName' is equal to "A"
# Then select only the 'id' and 'lastName' columns for those rows
# dataframe["firstName"] == "A"  A boolean condition that returns True for rows where the first name is "A"
# ["id", "lastName"]  Specifies which columns to return from those rows.
# Before the comma, specify the condition (firstname =A); after the comma, put what columns you are going to read

dataframe.loc[dataframe["firstName"] == "A",["id","lastName"]]
```

```python
#display id and lastname when firstName not equals to A

dataframe.loc[dataframe["firstName"] != "A", ["id","lastName"]]
```

```python
#display the rows with firstName not equals to A

dataframe.loc[dataframe["firstName"] != "A", ]
```

```python
# Define two boolean conditions:
# condition1 is True where 'firstName' is "A"
# condition2 is True where 'lastName' is "A"

condition1 = dataframe["firstName"] == "A"
condition2 = dataframe["lastName"] == "A"

# Select rows where both conditions are True (firstName is "A" AND lastName is "A")
dataframe.loc[(condition1 & condition2)]
```

## or

```python
# Select rows where either condition is True (firstName is "A" OR lastName is "A")

dataframe.loc[(condition1 | condition2),]
```

```python
#display id and lastname when either condition 1 or 2 = A

dataframe.loc[(condition1 | condition2), ["id","lastName"]]
```

if  condition:
    sequence of statements

if the condition is true, then goes to the sequences of statements. Otherwise, skip it.


```python
marks = input("please enter your marks: ")
```

```python
mark1 = int(marks)
mark1
```

1. These statements consist of a condition and just a single sequence of statements

2. A Boolean value is a result of a comparison expression. These expressions includes the following relational
operators: ==; ! =; <; >;<=;>=

3. If the condition is True, the sequence of statements is executed. Otherwise, control proceeds to the next statement outside the selection statement

## Two-way Selection Statements - If: Else:

Each <sequence of statements> must be indented at least one space beyond both symbols 'if' and 'else'
If the condition is True, the first sequence of statements is executed. Otherwise, the second sequence of statements is executed. Then control proceeds to the next statement outside the selection statement

```python
numericGrade  = int(input("Enter the numeric grade"))
if numericGrade >= 50:
    print("pass")
else:
    print("fail")
```

```python
password = input("Enter password: ")
if password == 'qwer1234':
    print("correct password")
else:
    print("Incorrect Password")
```

## Multi-way Selection Statements: Syntax and Structure

1. Multi-way statements consider each condition until one evaluates to True or they all evaluate to False
2. When a condition evaluates to True, the corresponding sequence of statements is executed and control skips to the end of the entire selection statement.
3. If no condition evaluates to True, then the corresponding <sequence of statements> to the trailing else is executed

```python
numericGrade  =  int(input("please enter your grade"))
if numericGrade >= 90:
    print("Grade A")
elif numericGrade >= 80:
    print("Grade B")
elif numericGrade >= 70:
    print("Grade C")
elif numericGrade >= 60:
    print("Grade D")
else :
    print("Fail")
```
